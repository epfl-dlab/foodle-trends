```python
import pandas as pd
import gtab
import os
import matplotlib.pyplot as plt
import time

t = gtab.GTAB()

parentDirectory = os.path.abspath(os.path.join(os.path.join(os.getcwd(), os.pardir), os.pardir))
DATA_DIR = parentDirectory +'/data/'
FIGURES_DIR = parentDirectory +'/figures/'
```

```python
df_codes = pd.read_csv(DATA_DIR+"country_codes_filtered.csv")
```

```python
selected_countries = ['FR','DE','US','IT','CA','GB','ES','AU','MX','BR','IN','DK']
```

```python
df_query = pd.read_csv(DATA_DIR+'wikidata_info_entities_cleaned_new2.csv')[['mid','name','class']]
```

```python
df_query = df_query.dropna()
```

```python
len(df_query)
```

```python
for country_code in selected_countries:
    
    cnt = 0
    for name, kg_id in zip(df_query.name.values, df_query.mid.values):
        cnt +=1
        
        '''
        if cnt >775 and country_code == "US":
            pass
        elif country_code != "US":
            pass
        else:
            continue
        '''
        
        #print("="*15)
        print(name,',',cnt,'/',1409)
        if not os.path.exists("./results/{}_{}.tsv".format(country_code, name)):
            # set active gtab according to the code! these are already installed
            t.set_active_gtab("google_anchorbank_geo={}_timeframe=2019-01-01 2020-12-31.tsv".format(country_code))

            try:
                # makes the query
                df_r = t.new_query(kg_id)

                # saves the query
                df_r.to_csv("./results/{}_{}.tsv".format(country_code,name))
                #time.sleep(60)
                
                
            except (AttributeError, KeyError) as error:
                print('Skipped:',country_code,name)
                pd.DataFrame([]).to_csv("./results/{}_{}.tsv".format(country_code,name))
                #time.sleep(60)
                continue
        else:
            print("already exists!")
```
