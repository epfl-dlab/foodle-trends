```python
for country_code in selected_countries:
    for name, kg_id in zip(df_query.name.values, df_query.kg_id.values):

        print("="*15)
        print(name)
        if not os.path.exists("./results_modes/{}_{}.tsv".format(country_code, name)):
            # set active gtab according to the code! these are already installed
            t.set_active_gtab("google_anchorbank_geo={}_timeframe=2019-01-01 2020-12-31.tsv".format(country_code))

            try:
                # makes the query
                df_r = t.new_query(kg_id)

                # saves the query
                df_r.to_csv("./results/{}_{}.tsv".format(country_code,name))
            except AttributeError as error:
                print('Skipped:',country_code,name)
                continue
        else:
            print(name, country_code, "already exists!")
        
```
